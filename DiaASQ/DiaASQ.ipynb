```python
import os
import pandas as pd

File_path=os.path.join(os.getcwd(),'DiaASQ/data/dataset/jsons_en/train.json')

df= pd.read_json(File_path)
df.head(10)
```

```python
def assign_positions_to_paragraphs(paragraphs, triplets):
    result = []
    spot=0
    for p in paragraphs:
      assigned_targets = []
      for t in triplets:
          index = t[0]
          count = t[1]
          aspect = t[7]
          sentiment= t[6]
          category = t[8]
          opinion =t[9]
          count=count-index if count-index>0 else count
          if (index-spot) < len(p.split()) and (index-spot)>0:
            merged_elements = ' '.join(p.split()[index-spot:index-spot + count]).replace(',','').strip()
            new_s=p.split()

            del new_s[index-spot:index-spot + count]
            new_s.insert(index-spot, merged_elements)
            # Process the current paragraph
            try:
              if new_s.index(aspect)==(index-spot):
                assigned_targets.append([aspect,category,opinion,sentiment])
            except:
                assigned_targets.append(['NULL',category,opinion,sentiment])
          
      spot += len(p.split())
      # Append the paragraph with its assigned positions
      result.append([p, assigned_targets])
      assigned_targets = []
    return result

sentences = list(df.loc[0, 'sentences'])
triplets = list(df.loc[0, 'triplets'])

result = assign_positions_to_paragraphs(sentences,triplets)
for r in result:
   print (r)
```
